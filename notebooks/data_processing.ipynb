```python
import pandas as pd
import country_converter as coco
import numpy as np

from dotenv import load_dotenv

load_dotenv()
```

## Billionaires count

```python
# Import billionaires

df_billionaires_birth = pd.read_csv("C:/Users/keita/git/DataVis/data/raw/Country_of_birth.csv", usecols=["Country of birth"])
df_billionaires_all = pd.read_csv("C:/Users/keita/git/DataVis/data/raw/Billionaires Statistics Dataset.csv")
df_billionaires_all = df_billionaires_all[df_billionaires_all["selfMade"] == True]
df_billionaires_all.reset_index(drop=True, inplace=True)
df_billionaires_birth.reset_index(drop=True, inplace=True)
merged_df = pd.concat([df_billionaires_all, df_billionaires_birth], axis=1)
merged_df
```

```python
# to csv for used billionaires
merged_df.to_csv("C:/Users/keita/git/DataVis/data/processed/billionaires.csv", index=False)
```

```python
# Compute number of billionaires

merged_df = merged_df[merged_df["Country of birth"] == merged_df["country"]]
df_grouped = merged_df.groupby('Country of birth').agg(
    billionaires_count=('personName', 'count'),
)
df_grouped = df_grouped.reset_index()
df_grouped["Country of birth"] = df_grouped["Country of birth"].apply(lambda x: coco.convert(names=x, to='name_short'))

df_grouped.rename(columns={'Country of birth': 'country'}, inplace=True)
df_grouped
```

## Basic data from billionaires dataset about countries

```python
# Import billionaires country

billionaires = pd.read_csv("C:/Users/keita/git/DataVis/data/raw/Billionaires Statistics Dataset.csv")
df = billionaires.groupby('country').agg(
    latitude=('latitude_country', 'first'),
    longitude=('longitude_country', 'first')
)
df.index = df.index.map(lambda country: coco.convert(names=country, to='name_short'))
df.reset_index(inplace=True)
df
```

```python
# Merge two df and df_gdp

df = pd.merge(df, df_grouped, on='country', how='inner')
df
```

## Gini

```python
# Import Gini HDI

df_gini = pd.read_csv("C:/Users/keita/git/DataVis/data/raw/HDR23-24_HDI_Table.csv", usecols=["Country", "HDI_2022", "gini_2010-2022"], decimal = ',')
df_gini.dropna(subset=["HDI_2022", "gini_2010-2022"], inplace=True)
df_gini.rename(columns = {'Country':'country', 'HDI_2022':'HDI', 'gini_2010-2022':'Gini'}, inplace=True)
df_gini["Gini"] = df_gini["Gini"].replace("..", np.nan)
df_gini["Gini"] = df_gini["Gini"].str.replace(",", ".").astype(float)
df_gini['country'] = df_gini['country'].apply(lambda x: coco.convert(names=x, to='name_short'))
df_gini

```
